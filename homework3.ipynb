```r
A <- subset(titanicnew, Survived == 1 & titanicnew$Fare > mean(titanicnew$Fare))
B <- subset(titanicnew, Survived == 0 & titanicnew$Fare > mean(titanicnew$Fare))
length(A$Fare) / length(B$Fare)
```

```r
sfare <- (c(titanicnew$Fare) - mean(titanicnew$Fare))/ sd(titanicnew$Fare)
titanicnewest <- data.frame(titanicnew, sfare)
```

```r
youngerthanaverage <- subset(titanicnewest, titanicnewest$Age < mean(titanicnewest$Age))
ggplot(data=youngerthanaverage) + geom_bar(mapping = aes(x = Sex, fill = Survived))
```

```r
survivedandpaidmore <- subset(titanicnewest, Survived == 1 & Fare > mean(titanicnewest$Fare))
ggplot(data = survivedandpaidmore) + geom_density(mapping = aes(x = Age))
```

```r
cfare <- titanicnew$Fare
cfare[cfare < mean(cfare)] <- "Cheap"
titaniccfare <- data.frame(titanicnew, cfare)
```

```r
cage <- titanicnew$Age
cage[cage <= 10 & cage >= 0] <- 0
cage[cage <= 20 & cage >= 10] <- 1
cage[cage <= 30 & cage >= 20] <- 2
cage[cage <= 40 & cage >= 30] <- 3
cage[cage <= 50 & cage >= 40] <- 4
cage[cage <= 60 & cage >= 50] <- 5
cage[cage <= 70 & cage >= 60] <- 6
cage[cage <= 80 & cage >= 70] <- 7
cage[cage <= 90 & cage >= 80] <- 8
cage[cage <= 100 & cage >= 90] <- 9
cage[cage <= 110 & cage >= 100] <- 10
titaniccage <- data.frame(titanicnew, cage)
```

```r
titaniccage <- titaniccage[titaniccage$Fare >= mean(titaniccage$Fare),]
ggplot(data = titaniccage) + geom_bar(mapping = aes(x = cage, fill = factor(Survived)))
```
